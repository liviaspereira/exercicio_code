Achando a *média*, *mediana* e *moda* de horas de sono para cada uma das profissões:

```python
# Agrupar por profissão
grupo_profissao = df1.groupby('Profissão')['Duração do sono']

# Calcular média, mediana e moda para cada profissão
media_por_profissao = grupo_profissao.mean()
mediana_por_profissao = grupo_profissao.median()
moda_por_profissao = grupo_profissao.apply(lambda x: x.mode().iloc[0] if not x.mode().empty else None)

# Criar um novo DataFrame com os resultados
resultados = pd.DataFrame({
    'Média': media_por_profissao,
    'Mediana': mediana_por_profissao,
    'Moda': moda_por_profissao
}).reset_index()

print(resultados)
```

Das pessoas que atuam com engenharia de software qual a porcentagem de obesos:

```python
# Filtrar o DataFrame para a profissão "Eng. de Software"
eng_software_df = df1[df1['Profissão'] == 'Eng. de Software']

# Calcular a porcentagem de "obesidade" na profissão "Eng. de Software"
porcentagem_obesidade = (eng_software_df['Categoria IMC'].value_counts(normalize=True) * 100).get('Obesidade', 0)

print(f'A porcentagem de "Obesidade" na profissão "Eng. de Software" é: {porcentagem_obesidade:.2f}%')
```

De acordo com os dados, advogar ou ser representante de vendas faz você dormir menos? (Use o método ‘isin’, considere a média)

```python
# Calcular a média da duração do sono para cada profissão
media_duracao_sono = df1.groupby('Profissão')['Duração do sono'].mean()

# Ordenar as profissões com base na média da duração do sono em ordem crescente
profissoes_menos_dormem = media_duracao_sono.sort_values().head(2)

print("As duas profissões que menos dormem são:")
print(profissoes_menos_dormem)
```

```python
# Ordenar as profissões com base na média da duração do sono em ordem decrescente
profissoes_mais_dormem = media_duracao_sono.sort_values(ascending=False).head(2)
print("As duas profissões que menos dormem são:")
print(profissoes_mais_dormem)
```

Entre quem fez enfermagem e quem fez medicina, quem tem menos horas de sono? (Use o método ‘isin’, considere a média)

```python
# Filtrar o DataFrame para incluir apenas 'Enfermagem' e 'Medicina'
profissoes_interessantes1 = ['Enfermeiro(a)', 'Médico(a)']
filtro = df1['Profissão'].isin(profissoes_interessantes1)
df_filtrado = df1[filtro]

# Calcular a média da duração do sono para cada profissão
media_duracao_sono = df_filtrado.groupby('Profissão')['Duração do sono'].mean()

# Comparar as médias
if media_duracao_sono['Enfermeiro(a)'] < media_duracao_sono['Médico(a)']:
    print("Quem fez Enfermagem tem menos horas de sono.")
elif media_duracao_sono['Enfermeiro(a)'] > media_duracao_sono['Médico(a)']:
    print("Quem fez Medicina tem menos horas de sono.")
else:
    print("Quem fez Enfermagem e Medicina têm a mesma quantidade de horas de sono.")
```

Achar a profissão menos frequente:

```python
# Usar value_counts para contar a frequência de cada profissão
contagem_profissoes = df1['Profissão'].value_counts()

# Encontrar a profissão menos frequente
profissao_menos_frequente = contagem_profissoes.idxmin()

print(f'A profissão menos frequente é: {profissao_menos_frequente}')
```

Quem tem maior pressão sanguínea média, homens ou mulheres? (Considere a média)

```python
# Extrair os valores antes da barra e converter para inteiros
df1['Pressao_antes_da_barra'] = df1['Pressão Sanguínea'].str.extract('(\d+)').astype(int)

# Calcular a média da pressão sanguínea para homens e mulheres
media_pressao_por_genero = df1.groupby('Gênero')['Pressao_antes_da_barra'].mean()

# Comparar as médias
if media_pressao_por_genero['Homem'] > media_pressao_por_genero['Mulher']:
    print("Homens têm uma pressão sanguínea média maior.")
elif media_pressao_por_genero['Homem'] < media_pressao_por_genero['Mulher']:
    print("Mulheres têm uma pressão sanguínea média maior.")
else:
    print("Homens e mulheres têm a mesma pressão sanguínea média.")

# Encontrar a maior pressão sanguínea antes da barra
maior_pressao_antes_da_barra = df1['Pressao_antes_da_barra'].max()

print(f'A maior pressão sanguínea antes da barra é: {maior_pressao_antes_da_barra}')
```

É predominante entre os participantes dormir 8 horas por dia (considere usar Moda como medida)?

```python
# Calcular a moda da duração do sono
moda_duracao_sono = df1['Duração do sono'].mode()

# Verificar se 8 horas é a moda (valor mais frequente)
if 8 in moda_duracao_sono.values:
    print("Dormir 8 horas por dia é predominante.")
else:
    print("Dormir 8 horas por dia não é predominante.")
```

 Pessoas com frequências cardíacas acima de 70 dão mais passos que
pessoas com frequência cardíaca menor ou igual a 70? (Use a média)

```python
# Criar uma coluna indicando se a frequência cardíaca é maior que 70 ou não
df1['Frequencia_Maior_70'] = df1['Frequência cardíaca'] > 70

# Calcular a média de passos para cada grupo
media_passos_por_frequencia = df1.groupby('Frequencia_Maior_70')['Passos diários'].mean()

# Comparar as médias
if media_passos_por_frequencia[True] > media_passos_por_frequencia[False]:
    print("Pessoas com frequências cardíacas acima de 70 dão mais passos.")
elif media_passos_por_frequencia[True] < media_passos_por_frequencia[False]:
    print("Pessoas com frequências cardíacas 70 ou abaixo dão mais passos.")
else:
    print("Pessoas com frequências cardíacas acima de 70 dão o mesmo número de passos que as de 70 ou abaixo.")
```
